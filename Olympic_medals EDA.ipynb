```python
country_medalcount = data.groupby(['Year','Country'])['Medal'].count().reset_index()
#print(country_medalcount)
country_medalcount.sort_values(by='Year')
l =list(country_medalcount.Country.unique())
ratio = []
for i in list(country_medalcount.Country.unique()):
    x = country_medalcount[country_medalcount["Country"] == i]
    #print (x)
    medals = sum(x.Medal)/len(x)
    #print(medals)
    ratio.append(medals)

medal_ratio = pd.DataFrame({"country": l,
    "ratio": ratio})

#print (medal_ratio)
medal_ratio = medal_ratio.sort_values(by= 'ratio', ascending = False)
```

```python
plt.figure(figsize = (10,10))
sns.barplot( y = 'country', x = 'ratio', data = medal_ratio[:20])
```

*For example, having 500 medals when you participated 2 times is not the same if you have participated 5 times. Ratio would be totally different. Above Graph gives us the value which informs on how many medals received and how many times Country participated in the Olympic Game.*

```python
data1 = pd.get_dummies(data, columns = ['Medal'])
country_list = list(data1['Country'].unique())
#print(country_list)
medal_gold = []
medal_silver = []
medal_bronze = []
ratio_gold = []
ratio_silver = []
ratio_bronze = []
for i in country_list:
    x = data1[data1["Country"] == i]
    #print('---------------------------')
    #print(x.shape)
    gold = x.Medal_Gold.sum()
    medal_gold.append(gold)
    ratio_gold.append((gold)/len(x))
    #ratio_gold = ratio_gold.append(sum(x.Medal_Gold)/len(x))
    silver = x.Medal_Silver.sum()
    medal_silver.append(silver)
    ratio_silver.append(silver/len(x))
    bronze = x.Medal_Bronze.sum()
    medal_bronze.append(bronze)
    ratio_bronze.append(bronze/len(x))

medal_list = pd.DataFrame({"country": country_list,"medal_gold": medal_gold, 'medal_silver': medal_silver,'medal_bronze': medal_bronze}) 
medal_list['medal_total'] = medal_list.medal_gold + medal_list.medal_silver + medal_list.medal_bronze
medal_list = medal_list.sort_values(by="medal_total", ascending=False)
print (medal_list[:20])

ratio_list = pd.DataFrame({"country": country_list,"Ratio_gold_medal": ratio_gold, 'ratio_silver_medal': ratio_silver,
    'ratio_bronze_medal': ratio_bronze})

sorted_ratio_list = ratio_list[:20].sort_values(by= "Ratio_gold_medal", ascending = False)
```

*Above value represents the count of different medals each country received and its total count.*

```python
plt.figure(figsize=(15,3))
plt.xticks(rotation=90)

sns.lineplot(data = sorted_ratio_list[:20], x = 'country', y = 'Ratio_gold_medal')
sns.lineplot(data = sorted_ratio_list[:20], x = 'country', y = 'ratio_silver_medal')
sns.lineplot(data = sorted_ratio_list[:20], x = 'country', y = 'ratio_bronze_medal')
plt.legend(["Silver", 'Gold', 'Bronze'])
    
#chart= sns.countplot(x ='Sport', hue= 'Medal',data = data, order = data['Sport'].value_counts().index)
#chart.set_xticklabels(chart.get_xticklabels(), rotation=90)    
#chart.set_xticklabels(chart.get_xticklabels(), rotation = 90)
#sns.lineplot(ratio_list['Ratio_silver_medal'])
#sns.lineplot(ratio_list['Ratio_gold_medal'])
```
